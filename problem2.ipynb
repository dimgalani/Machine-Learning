Αποθηκεύστε τις μετρικές στις μεταβλητές r2_squared_error και r2_poisson αντίστοιχα. (1 μονάδα)

**4** Αναπτύξτε κώδικα με βάση την ιδιότητα [tree_](https://scikit-learn.org/stable/auto_examples/tree/plot_unveil_tree_structure.html#sphx-glr-auto-examples-tree-plot-unveil-tree-structure-py) του παρακάτω μοντέλου, προκειμένου να εμφανίσετε το μονοπάτι που ακολουθήθηκε στο δένδρο για την πρόβλεψη της πρώτης περίπτωσης του συνόλου ελέγχου. Σε περίπτωση που ελέγχεται δεύτερη φορά μια μεταβλητή στο μονοπάτι, θα πρέπει να διατηρούμε τον τελευταίο και στενότερο περιορισμό. Ακολουθούν παραδείγματα ορθής λειτουργίας, στα οποία φαίνεται και η μορφή αν-τότε που πρέπει να έχει το μονοπάτι. (4 μονάδες)

```python
from sklearn.tree import export_text
model = DecisionTreeRegressor(max_depth=3, random_state=42)
model.fit(X_train, y_train)
```

```python
def get_path(tree, instance, feature_names):
    feature = tree.tree_.feature
    threshold = tree.tree_.threshold
    value = tree.tree_.value

    # Initialize variables
    current_node = 0
    path_conditions = {}

    # Traverse the tree to find the path for the instance
    while True:
        if feature[current_node] != -2:  # If not a leaf node
            # Get the feature name and threshold for the current node
            feature_index = feature[current_node]
            feature_name = feature_names[feature_index]
            threshold_value = threshold[current_node]

            # Determine the direction to move in the tree
            if instance[feature_index] <= threshold_value:
                direction = "<="
            else:
                direction = ">"

            # Add or update the condition for the current feature
            if feature_index in path_conditions:
                # If the feature has been encountered before
                prev_feature_name, prev_direction, prev_threshold_value = path_conditions[feature_index]
                if prev_direction == direction:
                    # If the directions are the same, update the condition only if the new threshold is stricter
                    if direction == "<=":
                        if threshold_value < prev_threshold_value:
                            path_conditions[feature_index] = (feature_name, direction, threshold_value)
                    else:
                        if threshold_value > prev_threshold_value:
                            path_conditions[feature_index] = (feature_name, direction, threshold_value)
                else:
                    # If the directions are different, keep both conditions
                    path_conditions[feature_index + 0.5] = (feature_name, direction, threshold_value)
                    path_conditions[feature_index] = (prev_feature_name, prev_direction, prev_threshold_value)
            else:
                # If it's the first encounter of the feature, add it to the conditions
                path_conditions[feature_index] = (feature_name, direction, threshold_value)

            # Move to the left or right child node
            if direction == "<=":
                current_node = tree.tree_.children_left[current_node]
            else:
                current_node = tree.tree_.children_right[current_node]
        else:
            # Reached a leaf node, add the predicted value and exit the loop
            predicted_value = value[current_node][0][0]
            if predicted_value == int(predicted_value):
                formatted_predicted = f"{int(predicted_value)}"
            elif int(predicted_value * 10) == predicted_value * 10:
                formatted_predicted = f"{predicted_value:.1f}"
            else:
                formatted_predicted = f"{predicted_value:.2f}"
            
            # Construct the path string using the last and stricter condition for each feature
            path_conditions_str = [f"{name} {direction} {value:.2f}" for _, (name, direction, value) in path_conditions.items()]
            path_conditions_str.append(f"τότε {formatted_predicted}")
            break
        
    return "αν " + " και ".join(path_conditions_str[:-1]) + " " + path_conditions_str[-1]

print(get_path(model, X_test[0], diabetes.feature_names))
print(get_path(model, X_test[1], diabetes.feature_names))
print(get_path(model, X_test[6], diabetes.feature_names))
```

```python
"""Τεστ ορθού υπολογισμού του μονοπατιού"""
assert get_path(model, X_test[0], diabetes.feature_names) == 'αν bmi <= 0.01 και s5 > 0.01 και s4 <= 0.09 τότε 159.57'
assert get_path(model, X_test[1], diabetes.feature_names) == 'αν bmi > 0.01 και bmi <= 0.07 και s6 <= 0.03 τότε 175.8'
assert get_path(model, X_test[6], diabetes.feature_names) == 'αν bmi > 0.07 και s2 > 0.02 τότε 225.75'
```

**6** Υλοποιήστε συνάρτηση, η οποία δοθέντος ενός δενδρικού μοντέλου παλινδρόμησης, διερευνά όλες τις ακέραιες τιμές της μεταβλητής min_samples_leaf από ένα κάτω όριο μέχρι ένα άνω όριο και επιστρέφει την βέλτιστη με βάση το mean squared error στο παραπάνω σύνολο επικύρωσης έπειτα από εκπαίδευση στο παραπάνω σύνολο εκπαίδευσης. Επιπλέον επιστρέφει και το mean squared error (στρογγυλοποιημένο σε ακέραιο) που πετυχαίνει στο σύνολο ελέγχου το μοντέλο με την βέλτιστη επίδοση στο σύνολο επικύρωσης. (2 μονάδες)

```python
from sklearn.metrics import mean_squared_error

def best_min_samples_leaf(model, min, max):
    best_leaf = None
    best_mse_val = float('inf')
    best_mse_test = None

    for min_samples in range(min, max + 1):
        # Set the min_samples_leaf parameter
        model.set_params(min_samples_leaf=min_samples)
        
        # Train the model on the training set
        model.fit(X_train, y_train)
        
        # Predict on the validation set
        y_val_pred = model.predict(X_val)
        
        # Calculate MSE on the validation set
        mse_val = mean_squared_error(y_val, y_val_pred)
        
        # Update the best parameters if the current MSE is lower
        if mse_val < best_mse_val:
            best_mse_val = mse_val
            best_leaf = min_samples
            
            # Predict on the test set with the current best model
            y_test_pred = model.predict(X_test)
            best_mse_test = mean_squared_error(y_test, y_test_pred)
    
    # Return the best min_samples_leaf and the rounded MSE on the test set
    return best_leaf, int(round(best_mse_test))
```

```python
"""Τεστ ορθής υλοποίησης συνάρτησης"""
model = DecisionTreeRegressor(max_depth=3, random_state=42)
assert best_min_samples_leaf(model, 1, 8) == (7, 3358)
assert best_min_samples_leaf(model, 5, 15) == (11, 3271)
```

**7** Υλοποιήστε συνάρτηση η οποία δοθέντος ενός μοντέλου παλινδρόμησης, ενός συνόλου δεδομένων και μίας μετρικής, το αξιολογεί με τη μέθοδο της σταυρωτής επικύρωσης "άφησε ένα εκτός". (1 μονάδα)

```python
from sklearn.model_selection import LeaveOneOut
import numpy as np

def leave_one_out(model, X, y, metric):
    # Initialize LeaveOneOut object
    loo = LeaveOneOut()
    scores = []

    # Loop over each train-test split
    for train_index, test_index in loo.split(X):
        # Split data
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Train the model
        model.fit(X_train, y_train)
        
        # Evaluate the model
        y_pred = model.predict(X_test)
        
        # Compute the specified metric and append to list
        score = metric(y_test, y_pred)
        scores.append(score)

    # Compute the average of the scores
    average_score = np.mean(scores)
    return average_score
```
